```python
import pandas as pd
```

```python
df = pd.read_excel("./data/online_retail.xlsx")

df.head(12)
```

```python
df.shape
```

```python
to_drop = ["Description", "StockCode"]

df = df.drop(to_drop, axis=1)

df.head(12)
```

```python
df.info()
```

```python
df["Country"].unique()
```

```python
import datetime as dt

def get_month_date(x):
    return dt.datetime(x.year, x.month, 1)


df["InvoiceMonth"] = df["InvoiceDate"].apply(get_month_date)

df.head(12)
```

```python
df["CohortMonth"] = df.groupby("CustomerID")["InvoiceMonth"].transform('min')

df.head(12)
```

```python
def get_date_elements(df, column):
    day = df[column].dt.day
    month = df[column].dt.month
    year = df[column].dt.year
    return day, month, year

_, InvoiceMonth, InvoiceYear = get_date_elements(df, "InvoiceMonth")
_, CohortMonth, CohortYear = get_date_elements(df, "CohortMonth")
```

```python
year_diff = InvoiceYear - CohortYear
month_diff = InvoiceMonth - CohortMonth
df["CohortIndex"] = year_diff*12 + month_diff+1

df.head(12)
```

```python
cohort_result = df.groupby(["CohortMonth", "CohortIndex"])["CustomerID"].nunique().reset_index()

cohort_table = cohort_result.pivot(index="CohortMonth", columns="CohortIndex", values="CustomerID")

cohort_table
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


cohort_table.index = pd.to_datetime(cohort_table.index)

cohort_table.index = cohort_table.index.strftime('%B %Y')

plt.figure(figsize=(21, 10))
sns.heatmap(cohort_table, annot=True)

plt.show()
```
